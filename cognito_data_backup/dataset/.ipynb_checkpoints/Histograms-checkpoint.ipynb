```python
import glob
file_list = glob.glob("./mrt/*.txt")
print "Total subject count (finished 30-day study) = " + str(len(file_list))
```

### Create data 

The following line of codes will go thorugh the directory where data is located. Then it will create a csv file 
called './stat111/hist_1.csv'. 

The columns of the csv file are the following:

- username
- day_count (1-30) 
- isDailySurveyCompleted (1=completed, 0=not completed) 
- isActiveTasksCompleted (1=completed, 0=not completed) 
- at_tapCount 
- at_spatialSec 
- ds_stress(0-4 likert) 
- ds_valence(-5 to 5) 
- ds_arousal(-5 to 5) 
- ds_FreeTime (minutes) 
- ds_Loneliness (0-4 likert) 
- ds_HowNewTheDayWas (0-4 likert) 
- ds_HowGoodTomorrowWillBeOptimism (0-4 likert) 

```python
import json
from pprint import pprint
from datetime import datetime, timedelta
import pdb

#initialize count
counts_over_days = []
for i in range(30):
    counts_over_days.append(0)

total_adherence_subject = []
for j in range(len(file_list)):
    total_adherence_subject.append(0)
    

    
data_for_subject = ""    
for j in range(len(file_list)):    
    data = json.load(open(file_list[j]))
    daily_survey = data[u'survey_data'][u'daily_survey']
    daily_survey_keys = sorted(daily_survey.keys()) #this will sort alphabatically
    active_tasks_survey = data[u'survey_data'][u'active_tasks_survey']
    active_tasks_survey_keys = sorted(active_tasks_survey.keys()) #this will sort alphabatically
     #this will sort alphabatically  
    username = str(data[u'username'])
    
    # print username
    if len(daily_survey_keys) == 0:
        continue
        
    life_insights_data = data[u'life-insights']    
    first_date = daily_survey_keys[0]
    
    next_date = first_date
    username = "s" + str(j)

    for i in range(30):
        date = datetime.strptime(next_date, "%Y%m%d")
        data_for_subject = data_for_subject + username + "," + str(i+1) # + "," + next_date
        
        
        #structure of the data
        # username, day_count (1-30), isDailySurveyDone, isActiveTasksDone, at_tapCount, at_spatialSec
        # ds_stress(0-4 likert), ds_valence(-5 to 5), ds_arousal(-5 to 5), ds_FreeTime (minutes), ds_Loneliness (0-4 likert), 
        #  ds_HowNew (0-4 likert), ds_HowGoodTomorrowWillBeOptimism (0-4 likert) 
        
        # at_tapCount
        if next_date in daily_survey:
            data_for_subject = data_for_subject  + ",1"
        else:
            data_for_subject = data_for_subject  + ",0"
            
            
        # isActiveTasksDone
        if next_date in active_tasks_survey:
            data_for_subject = data_for_subject  + ",1"
        else:
            data_for_subject = data_for_subject  + ",0"
            
        
        # at_tapCount
        if next_date in life_insights_data["at_tap"]:
            data_for_subject = data_for_subject  + "," + str(life_insights_data["at_tap"][next_date])
        else:
            data_for_subject = data_for_subject  + ",NaN"
            
            
        # at_spatialSec
        if next_date in life_insights_data["at_sp"]:
            data_for_subject = data_for_subject  + "," + str(life_insights_data["at_sp"][next_date])
        else:
            data_for_subject = data_for_subject  + ",NaN"   
        
        
        
        # daily survey data
        if next_date in life_insights_data["daily_survey"]:
            daily_survey_data_str = life_insights_data["daily_survey"][next_date].replace(':', ',')
            daily_survey_data_str = daily_survey_data_str.replace('undefined', 'NaN')
            
            #
            daily_survey_data_tok = daily_survey_data_str.split(",")
            
            
            #daily_survey_data_tok_stress = int(daily_survey_data_tok[0])
            #daily_survey_data_tok_FreeTime = int(daily_survey_data_tok[3])
            #daily_survey_data_tok_Loneliness = int(daily_survey_data_tok[4])
            #daily_survey_data_tok_HowNew = int(daily_survey_data_tok[5])
            #daily_survey_data_tok_HowGoodTomorrowWillBe = int(daily_survey_data_tok[6])
            
            
            
            #--- 
            if daily_survey_data_tok[1] != 'NaN':
                daily_survey_data_tok_valence = float(daily_survey_data_tok[1])
                daily_survey_data_tok_valence = str(daily_survey_data_tok_valence)
            else: 
                #means emotion was not reported.
                daily_survey_data_tok_valence = daily_survey_data_tok[1]
                daily_survey_data_tok_arousal = 'NaN'
                daily_survey_data_tok.append(daily_survey_data_tok[5])
                daily_survey_data_tok[5] = daily_survey_data_tok[4]
                daily_survey_data_tok[4] = daily_survey_data_tok[3]
                daily_survey_data_tok[3] = daily_survey_data_tok[2]
                
            #--- 
            if daily_survey_data_tok[2] != 'NaN':
                daily_survey_data_tok_arousal = float(daily_survey_data_tok[2]) + 5.0
                daily_survey_data_tok_arousal = str(daily_survey_data_tok_arousal)
            else:
                daily_survey_data_tok_arousal = daily_survey_data_tok[2]
                
            if len(daily_survey_data_tok) < 7:
                pdb.set_trace()
            
            daily_survey_data_str1 = daily_survey_data_tok[0] + ','
            daily_survey_data_str1 = daily_survey_data_str1 + daily_survey_data_tok_valence + ','
            daily_survey_data_str1 = daily_survey_data_str1 + daily_survey_data_tok_arousal + ','
            daily_survey_data_str1 = daily_survey_data_str1 + daily_survey_data_tok[3] + ','
            daily_survey_data_str1 = daily_survey_data_str1 + daily_survey_data_tok[4] + ','
            daily_survey_data_str1 = daily_survey_data_str1 + daily_survey_data_tok[5] + ','
            daily_survey_data_str1 = daily_survey_data_str1 + daily_survey_data_tok[6]
            
            #pdb.set_trace()
            #--- 
            data_for_subject = data_for_subject  + "," + str(daily_survey_data_str1)
            
        else:
            data_for_subject = data_for_subject  + ",NaN,NaN,NaN,NaN,NaN,NaN,NaN"
        
        
        #   
        data_for_subject = data_for_subject + "\n"
        
        

        #     
        modified_date = date + timedelta(days=1)
        next_date = datetime.strftime(modified_date, "%Y%m%d")

        
```

```python
file = open('./stat111/hist_1.csv', 'w')
file.write(data_for_subject)
file.close()
```

### Histogram 1: Whether daily surveys were completed
- 1=completed, 0=not completed
- probabilites are below the histograms

```python
f = open('./stat111/hist_1.csv', 'rb')
all_data = []
reader = csv.reader(f)
for row in reader:
    all_data.append(row)
f.close()


adherence_over_days = []
for i in range(30):
    data = []
    count_adhered = 0
    total_people = 0
    for j in range(len(all_data)):
        #print row
        row = all_data[j]
        day = int(row[1]) - 1
        if day == i:
            data.append(int(row[2]))
            if int(row[2]) == 1:
                count_adhered = count_adhered + 1.0
            total_people = total_people + 1.0
    adherence_over_days.append(count_adhered/total_people)
    
    
    #print data
    ii = i%3
    
    #size the plot
    if ii == 0:
        plt.figure(num=None, figsize=(12, 3), dpi=80, facecolor='w', edgecolor='k')
        
    plt.subplot(1, 3, ii+1)
    plt.hist(data, normed=False, bins=np.arange(-0.5, 2, 1))
    plt.xlim(-1, 2)
    plt.ylim(0, 37)
    plt.ylabel('Day ' + str(i+1));
    
    #start showing off
    if ii == 2:
        print "Day " + str(i-1) + ' to ' + str(i+1)
        plt.show()

```

```python
output_array = np.array(adherence_over_days)
np.savetxt("daily_survey_adherence.csv", output_array, fmt='%f', delimiter=",")
adherence_over_days #for daily surveys
```

## Active task: Tap count

```python
f = open('./stat111/hist_1.csv', 'rb')
all_data = []
reader = csv.reader(f)
for row in reader:
    all_data.append(row)
f.close()

min_q = 0
max_q = 150 #this is hard coded. Calculated form the histograms below
step_size = 15

all_q_bins = []
for i in range(30):
    data = []
    for j in range(len(all_data)):
        #print row
        row = all_data[j]
        day = int(row[1]) - 1
        if day == i:
            if row[4] == 'NaN':
                #data.append(-50)
                pass
            else:
                data.append(int(row[4]))
    #print data
    ii = i%3
    
    #size the plot
    if ii == 0:
        plt.figure(num=None, figsize=(12, 3), dpi=80, facecolor='w', edgecolor='k')
        
    plt.subplot(1, 3, ii+1)
    #plt.hist(data, normed=True, bins=np.arange(-0.5, 2, 1))
    plt.hist(data, normed=False)
    plt.xlim(0, 150)
    plt.ylim(0, len(data))
    plt.ylabel('Day ' + str(i+1));
    
    #start showing off
    if ii == 2:
        print "Day " + str(i-1) + ' to ' + str(i+1)
        plt.show()
        
    #
    q_bins = []
    q_bins_labels = []
    for ii in range(min_q,max_q,step_size):
        nn = 0
        for kk in range(len(data)):
            if (data[kk] >= ii) and (data[kk] < (ii+step_size)):
                nn = nn + 1
        q_bins.append(nn)
        q_bins_labels.append('>=' + str(ii) + " && " + '<' + str(ii+step_size))
    all_q_bins.append(q_bins)
```

```python
print "column names"
print q_bins_labels
np.set_printoptions(precision=2)
output_array = np.array(all_q_bins)
np.savetxt("at_tapcount.csv", output_array, fmt='%i', delimiter=",")
print "\nEach row is for one day. There are 30 rows."
all_q_bins
```

## Daily survey: How New and Exciting The Day Was (0-4 likert)

```python
f = open('./stat111/hist_1.csv', 'rb')
all_data = []
reader = csv.reader(f)
for row in reader:
    all_data.append(row)
f.close()


min_q = 0
max_q = 5 #this is hard coded. Calculated form the histograms below
step_size = 1

all_q_bins = []
for i in range(30):
    data = []
    for j in range(len(all_data)):
        #print row
        row = all_data[j]
        day = int(row[1]) - 1
        if day == i:
            if row[11] == 'NaN':
                #data.append(-50)
                pass
            else:
                data.append(int(row[11]))
    #print data
    ii = i%3
    
    #size the plot
    if ii == 0:
        plt.figure(num=None, figsize=(12, 3), dpi=80, facecolor='w', edgecolor='k')
        
    plt.subplot(1, 3, ii+1)
    #plt.hist(data, normed=True, bins=np.arange(-0.5, 2, 1))
    plt.hist(data, normed=False)
    plt.xlim(-1, 5)
    plt.ylim(0, len(data))
    plt.ylabel('Day ' + str(i+1));
    
    #start showing off
    if ii == 2:
        print "Day " + str(i-1) + ' to ' + str(i+1)
        plt.show()
        
    #
    q_bins = []
    q_bins_labels = []
    for ii in np.arange(min_q,max_q,step_size):
        nn = 0
        for kk in range(len(data)):
            if (data[kk] >= ii) and (data[kk] < (ii+step_size)):
                nn = nn + 1
        q_bins.append(nn)
        q_bins_labels.append(str(ii))
    all_q_bins.append(q_bins)
```

```python
print "column names"
print q_bins_labels
print "\nEach row is for one day. There are 30 rows."
all_q_bins
```
